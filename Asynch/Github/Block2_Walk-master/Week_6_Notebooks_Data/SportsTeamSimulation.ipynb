```python
# Sports Team Simulation
# Baseball
# Based on Miller (2015)

import numpy as np
import random
from scipy.stats import nbinom
from array import array

# import pyprind   # PROVIDES TIMING AWARENESS
```

```python
def simulator(home_mean, away_mean, niterations):
    # estimates probability of home team win
    seed(1234)  # set to obtain reproducible results
    home_game_score = [0] * niterations
    away_game_score = [0] * niterations
    home_win = [0] * niterations
    i = 0
    while (i < niterations):
        home_game_score[i] = \
            nbinom.rvs(n = 4.0, p = 4.0/(4.0 + home_mean), size = 1)[0] 
        away_game_score[i] = \
            nbinom.rvs(n = 4.0, p = 4.0/(4.0 + away_mean), size = 1)[0]         
        if (home_game_score[i] > away_game_score[i]):
            home_win[i] = 1
        if ((away_game_score[i] > home_game_score[i]) or \
            (away_game_score[i] < home_game_score[i])):
            i = i + 1 
    n_home_win = sum(home_win)
    return n_home_win / niterations      
```

```python
import numpy as np
import random
import matplotlib.pyplot as plt

def football_simulator(home_mean, away_mean, niterations):
    # simulate a football game and estimate the probability of the home team winning
    np.random.seed(1234)  # set to obtain reproducible results
    home_score = np.random.poisson(home_mean, size=niterations)
    away_score = np.random.poisson(away_mean, size=niterations)
    home_win = (home_score > away_score).astype(int)
    n_home_win = np.sum(home_win)
    return n_home_win / niterations

# set the scoring rates for each team
home_mean = 21
away_mean = 13

# simulate the game and estimate the probability of the home team winning
niterations = 10000
probmat = np.zeros((4, 4))  # create a matrix to store the probabilities
for home_score in range(4):
    for away_score in range(4):
        probmat[home_score, away_score] = football_simulator(home_mean + home_score, away_mean + away_score, niterations)

# create a probability heat-map using Matplotlib
fig, ax = plt.subplots()
im = ax.imshow(probmat, cmap='coolwarm', vmin=0, vmax=1)
ax.set_xticks(np.arange(4))
ax.set_yticks(np.arange(4))
ax.set_xticklabels(np.arange(4) + home_mean)
ax.set_yticklabels(np.arange(4) + away_mean)
plt.xlabel('Home team score')
plt.ylabel('Away team score')
plt.colorbar(im)
plt.show()
```
